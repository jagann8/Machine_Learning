## Apply method

```python
def add_rows(row):
  return row['A']+row['B']+row['C']


def main():
  data = {
      'A' : [1, 2, 3],
      'B' : [4, 5, 6],
      'C' : [7, 8, 9]
  }

  df = pd.DataFrame(data)
  print('Orginal DataFrame : \n', df)

  # apply the user defined function in every data frame
  df['add'] = df.apply(add_rows, axis=1)
  print('After applying function\n')
  print(df)


main()
```

```python
## by using lambda function in apply method

def add(A, B, C):
  return A+B+C


data = {
    'A' : [1,2,3],
    'B' : [4,5,6],
    'C' : [7,8,9]
}

dff = pd.DataFrame(data)

print('Orginal DataFrame : \n', dff)

dff['add'] = dff.apply(lambda row : add(row['A'], row['B'], row['C']), axis=1)
print('After apply method in dataframe : \n')
print(dff)

```
