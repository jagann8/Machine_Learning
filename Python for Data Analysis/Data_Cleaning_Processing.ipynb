```python
# Categorical extension type in pandas
fruits=['apple','orange','apple','apple']*2
N=len(fruits)
rng=np.random.default_rng(seed=12345)
data=pd.DataFrame({'fruits':fruits,
                   'basket_id':np.arange(N),
                   'count':rng.integers(3,15,size=N),
                   'weight':rng.uniform(0,4,size=N)
                   },columns=['fruits','basket_id','count','weight'])
data
```

```python
# Computational with categorical data
rng=np.random.default_rng(seed=12344)
draws=rng.standard_normal(1000)
draws[:5]
```

```python
bins=pd.cut(draws,4)
bins
```

```python
# here we adding labels to for better understand
bins=pd.qcut(draws,4,labels=['Q1','Q2','Q3','Q4'])
bins
```

```python
bins.codes[:10]
```

```python
bins=pd.Series(bins,name='quartile')
results=(pd.Series(draws).groupby(bins).agg(['count','min','max']).reset_index())
results
```

```python
results['quartile']
```

```python
# Better performance with categorical
N=10_00_000
labels=pd.Series(['foo','bar','baz','qux']*(N//4))
# now labels converted into categories
categories=labels.astype('category')
categories
```

```python
# to calculate memory capacity for both labels and categories
label_memory=labels.memory_usage(deep=True)
label_memory
```

```python
categories_memory=categories.memory_usage(deep=True)
categories_memory
```

```python
%time _ = labels.astype('category')
%time _
```

```python
%timeit labels.value_counts()
```

```python
%timeit categories.value_counts()
```

```python
# categorical Methods
s=pd.Series(['a','b','c','d']*2)
cat_s=s.astype('category')
cat_s.cat.categories
```

```python
cat_s.cat.codes
```

```python
actual_categories=['a','b','c','d','e']
cat_s2=cat_s.cat.set_categories(actual_categories)
cat_s2
```

```python
cat_s.value_counts()
```

```python
cat_s2.value_counts()
```

In large datasets, categoricals are often used as a convenient tool for memory savings and better performance. After you filter a large DataFrame or Series, many of the categories may not appear in the data. To help with this, we can use the remove_unused_categories method to trim unobserved categories:

```python
cat_s3=cat_s[cat_s.isin(['a','b'])]
cat_s3
```

```python
cat_s3.cat.remove_unused_categories()
```
