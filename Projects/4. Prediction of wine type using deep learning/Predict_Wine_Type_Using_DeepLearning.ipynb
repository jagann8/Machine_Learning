#### Dataset columns descriptions

1. Fixed acidity : the total acidity devided into two groups. the volatile acids and non volatile or fixed acids. these values represented by gm/dm3 in the datasets.

2. volatile acidity: the volatile acidity process of wine into vinegar. in this dataset, the volatile acidity represents gm/dm3

3. Citric acid : citric acid is one of the fixed acids in wine. it expressed as G/dm3

4. Residual sugar : residual sugar is sugar remaining after fermentation stops or is stopped. it is expressed by g/dm3 in the datasets.

  4.11 fermentation meaning
  -
  (a process of chemical change in food or drink because of the action of yeast or bacteria, which may cause it to produce bubbles or heat, or turn sugars in it into alcohol: Ethanol is an alcohol produced by fermentation of sugars.)

5. Chlorides : it is important contributor to saltiness in wine. the value of this variable is represented by gm/dm3 in the datasets.

6. Free sulpher dioxide: It is the part of sulpher dioxide that is added to the wine. the value of this variable represented by gm/dm3 in the datasets.

7. Total Sulpher dioxide: it is sum of the bound and free sulpher dioxide. the value of this varible represented by the gm/dm3 in the datasets.

```python

# since we have red and white wines are separate. so we want to combine these two datasets.
# so for red wine as 1 and white wine as 0

red['type'] = 1

white['type'] = 0

frames = [red ,white]
wines = pd.concat(frames)
```

```python
wines['type'].value_counts()
```

```python

X = wines.iloc[:,:-1].values
y = wines.iloc[:,-1].values

# splitting dataset into train set and validation data set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.34, random_state=45)
```

## Structure of the network

```python

# import the Sequential from keras models
from keras.models import Sequential

# import Dense from keras.layers
from keras.layers import Dense

# initialize the constructor
model = Sequential()

# add the input layer
model.add(Dense(13, activation = 'relu', input_shape=(12, )))

# add the hidden layer
model.add(Dense(9, activation = 'relu'))

# add the output layer
model.add(Dense(1, activation = 'sigmoid'))

# Model output_shape
model.output_shape

# model summary
model.summary()

# model config
model.get_config()

# list of the all tensors
model.get_weights()

# compile the model
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

## Training and Prediction

```python
from sklearn.metrics import accuracy_score

# training the model
model.fit(X_train, y_train, epochs=3, batch_size=1, verbose=1)

# predicting the accuracy
y_pred = model.predict(X_test)
print(y_pred)
```

The model predicted good accuracy. Hence our model trained well and able to predict the new data.
