```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data without feature scaling
data = {
    'Age': [25, 30, 35, 40, 45],
    'Income': [40, 50, 60, 70, 80],
    'Spending_Score': [45, 50, 60, 75, 80]
}

df = pd.DataFrame(data)

# Separate features and target
X = df[['Age', 'Income']]
y = df['Spending_Score']

# Create and fit a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict and plot the results
y_pred = model.predict(X)

plt.scatter(y, y_pred)
plt.xlabel('Actual Spending Score')
plt.ylabel('Predicted Spending Score')
plt.title('Regression without Feature Scaling')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Sample data with feature scaling (standardization)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create and fit a linear regression model
model = LinearRegression()
model.fit(X_scaled, y)

# Predict and plot the results
y_pred_scaled = model.predict(X_scaled)

plt.scatter(y, y_pred_scaled)
plt.xlabel('Actual Spending Score')
plt.ylabel('Predicted Spending Score')
plt.title('Regression with Feature Scaling (Standardization)')
plt.show()
```
