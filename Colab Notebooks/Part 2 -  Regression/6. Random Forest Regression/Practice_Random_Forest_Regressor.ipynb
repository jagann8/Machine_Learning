```python
# import the library

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import the dataset

dataset = pd.read_csv("/content/drive/MyDrive/Cloud Lab/Lab Experiments/Colab Notebooks/Part 2 -  Regression/6. Random Forest Regression/Position_Salaries.csv")
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
# train the whole dataset into the random forest regressor model

from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X, y)
```

```python
regressor.predict([[6.5]])
```

```python
# visualize the random forest regressor in high resolution with smooth curve

X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('RandomForestRegressor in high resolution with smooth curve')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
