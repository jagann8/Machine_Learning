- Decision Tree Regressor is not suitable for in one feature and two dimensional data.

- Decision tree regressor will better perform more than one feature.

```python
# import the dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Cloud Lab/Lab Experiments/Colab Notebooks/Part 2 -  Regression/5. Decision Tree Regression/Copy of Position_Salaries.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
# train the whole dataset into the decision tree regressor model

from sklearn.tree import DecisionTreeRegressor

regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(X ,y)
```

```python
# predict the new results

regressor.predict([[6.5]])
```

```python
# visualize the decision tree regressor in high resolution with smooth curve

X_grid = np.arange(min(X), max(X),0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid),color='blue')
plt.title('DecisionTreeRegressor Results in high resolution with smooth curve')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```
