```python
import pandas as pd
import numpy as np
```

```python
# import the dataset

dataset = pd.read_csv('/content/drive/MyDrive/Cloud Lab/Lab Experiments/Colab Notebooks/Part 3 - Classification/1. Logistic Regression/Copy of Social_Network_Ads.csv')

X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

```python
# splitting the dataset into train set and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
# now we doing the feature scaling for independent variables

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_test)
```

```python
print(X_train)
```

```python
# train the logistic regression model

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=0)
lr.fit(X_train, y_train)
```

```python
lr.predict([[19,40000]])
```

```python
y_pred = lr.predict(X_test)
print(y_pred)
```

```python
# checking the prediction results and actual results

np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)
```
